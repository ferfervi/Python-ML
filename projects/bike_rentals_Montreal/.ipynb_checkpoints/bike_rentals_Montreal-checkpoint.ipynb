For a same bike-line we will stufy differents parameters for rolling-mean [1 to 100] with intervals of 20 units (5 diferent param)

```python

for i in range(1,100,20):
        pd.rolling_mean(clean_df['Berri1'],i).plot(figsize=(15,6))
```

####  Which are the Bike-lines that are used more on daily a basis and which one are used most on the weekends?

```python
dow_df = clean_df.copy()
#create a new column
dow_df['dow'] = dow_df.index.dayofweek
dow_df.head()
```

Create a boolean variable to know if it's a week day or not

```python
dow_df['is_weekday'] = dow_df['dow'] <= 4
dow_df.head()
```

Let's count the number of rentals per week-days and weekend-days. To do so, we do a groupby operation and calculate the mean

```python
# Remove day number column since we have the is_weekday
dow_df = dow_df.drop('dow',1)
tday_df = dow_df.groupby('is_weekday').median()
```

```python
tday_df
```

Next, let's calcualte the percentages of each line per week days:

- List of bike-lines with more percentage (Descending order) of usage during the week

```python
# Viger and University are closer to 100% than the other ones, 
# so these two lines are more used on week days

path_tday_usage = tday_df.loc[True] / tday_df.sum()
path_tday_usage.order(ascending=False)
```

```python
#another method
path_tday_usage.sort_values(ascending = False)
```

- Viger is used more on week days that on weekends
- PierDup it is used similary on week days than on weekends

```python
path_tday_usage.sort_values(ascending = False).plot(kind='barh', figsize=(16,10))
```

```python
# Let's keep the 3 most used on week days , and the 3 more used on weekends
top_weekdays = path_tday_usage.sort_values(ascending=False)[:3]
top_weekends = path_tday_usage.sort_values()[:3]

# Since the previous result is a Series object with "index" the name of the 'rental-line' and
# with value the number of bike rentals, we can get the index to obtain these columns from the DF
clean_df[top_weekdays.index].plot(figsize=(15,6))
plt.legend(loc='center left', bbox_to_anchor=(1.0,0.5))

```

In the previous plot we are able to see for each rental-line how is ploted each week (ca 4 weeks per months) and how the the number of rentals decreases during the weeends, especially for the University line!! 

In the next plot let's analyze the traffic for in the lines that are more likely to have more traffic during the weekends (despite that in the best case may be similar to the traffic of the week days).

```python
clean_df[top_weekends.index].plot(figsize=(15,6))
```
